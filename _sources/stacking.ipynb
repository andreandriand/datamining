# Ensemble Stacking (Tugas 12)

Merupakan algoritma klasifikasi dengan mengambil output dari beberapa regresi dan klasifier model untuk digunakan sebagai input dari meta-classifier atau regressor

## Persiapan Environment

```python
import pandas as pd
data = pd.read_csv('https://raw.githubusercontent.com/andreandriand/dataset/main/credit_score.csv')
data
```

```python
data_without_column_for_convert = pd.DataFrame(data, columns = ['kode_kontrak','pendapatan_setahun_juta','durasi_pinjaman_bulan','jumlah_tanggungan','risk_rating'])
data_without_column_for_convert.head()
```

```python
data_ratarata=pd.get_dummies(data['rata_rata_overdue'])
data_ratarata.head()
```

```python
data_kpr_aktif=pd.get_dummies(data['kpr_aktif'])
data_kpr_aktif
```

## Data yang Belum Dinormalisasi

```python
data_new = pd.concat([data_without_column_for_convert, data_kpr_aktif,data_ratarata], axis=1)
data_new.head()
```

```python
data_risk_rating = pd.DataFrame(data, columns = ['risk_rating'])
data_risk_rating.head()
```

```python
data_without_risk_rating=data_new.drop(['risk_rating'], axis=1)
```

```python
data_new2 = pd.concat([data_without_risk_rating,data_risk_rating], axis=1)
data_new2
```

## Data yang telah Dinormalisasi

```python
from sklearn.preprocessing import MinMaxScaler
data_for_minmax_scaler=pd.DataFrame(data, columns = ['pendapatan_setahun_juta','durasi_pinjaman_bulan','jumlah_tanggungan'])
data_for_minmax_scaler.to_numpy()
scaler = MinMaxScaler()
data_hasil_minmax_scaler=scaler.fit_transform(data_for_minmax_scaler)
```

```python
data_hasil_minmax_scaler = pd.DataFrame(data_hasil_minmax_scaler,columns = ['pendapatan_setahun_juta','durasi_pinjaman_bulan','jumlah_tanggungan'])
data_hasil_minmax_scaler
```

```python
data_drop_column_for_minmaxscaler=data_new2.drop(['pendapatan_setahun_juta','durasi_pinjaman_bulan','jumlah_tanggungan'], axis=1)
data_drop_column_for_minmaxscaler
```

```python
data_new3 = pd.concat([data_drop_column_for_minmaxscaler,data_hasil_minmax_scaler], axis=1)
data_new3
```

```python
data_new3_without_risk_rating=data_new3.drop(['risk_rating'], axis=1)
data_new3_without_risk_rating
```

```python
data_new4 = pd.concat([data_new3_without_risk_rating,data_risk_rating], axis=1)
data_new4
```

```python
from sklearn.metrics import make_scorer, accuracy_score,precision_score
from sklearn.metrics import accuracy_score ,precision_score,recall_score,f1_score
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import KFold,train_test_split,cross_val_score
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.ensemble import StackingClassifier
X=data_new4.iloc[:,1:10].values
y=data_new4.iloc[:,11].values
estimators = [
    ('rf', RandomForestClassifier(n_estimators = 10)),
    ('gaussian', GaussianNB())
    
    ]
clf = StackingClassifier(estimators=estimators, final_estimator= GaussianNB())

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=0)

clf.fit(X_train, y_train).score(X_test, y_test)
```
