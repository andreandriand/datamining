##Menentukan Dissimilarity

```python
def diss(i,j):
  p = 2
  m = 0
  for col in ['category', 'started']:
    if data[col][i] == data[col][j]:
      m+=1
  return (p-m)/p
```

```python
print(diss(1,2))
print(diss(1,3))
print(diss(7,8))
print(diss(18,21))
```

##Numeric Attribute : Minkowski Distance

```python
def convert(s):
  return int("".join(s.split(",")))

def diss2(i,j,h):
  result = 0
  for col in ["subscribers", "video views", "video count"]:
    result = math.pow(abs(convert(data[col][i]) - convert(data[col][j])), h)
  return result**(1/h)

```

Manhattan Distance

```python
print(diss2(0,1,1))
print(diss2(0,2,1))
print(diss2(0,3,1))
```

Euclidean Distance

```python
print(diss2(0,1,3))
print(diss2(0,2,3))
print(diss2(0,3,3))
```
